Read a color image and display its reddish,greenisg and bluish image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("msELlJ2R_400x400.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
R, G, B = img[:,:,0], img[:,:,1], img[:,:,2]
```

```python
#red image
r_img = img.copy()
r_img[:,:,1], r_img[:,:,2] = 0, 0

plt.imshow(r_img)
```

```python
#green image
g_img = img.copy()
g_img[:,:,0], g_img[:,:,2] = 0, 0

plt.imshow(g_img)
```

```python
#blue image
b_img = img.copy()
b_img[:,:,0], b_img[:,:,1] = 0, 0

plt.imshow(b_img)
```
